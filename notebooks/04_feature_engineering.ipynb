```python
import pandas as pd

jobs = pd.read_csv("clean_jobs.csv")

print(jobs.shape)
```

```python
jobs["job_text"] = (
    jobs["title"].fillna("") + " " +
    jobs["description"].fillna("").str[:3000]
)
```

```python
jobs["job_text_length"] = jobs[
    "job_text"
].str.len()

jobs["job_text_length"].describe()
```

```python
for i in range(3):
    print(jobs["job_text"].iloc[i][:1000])
    print("="*100)
```

```python
jobs.to_csv(
    "feature_engineered_jobs.csv",
    index=False
)
```
