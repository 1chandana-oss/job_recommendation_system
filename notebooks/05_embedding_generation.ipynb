```python
import pandas as pd

jobs = pd.read_csv(
    "feature_engineered_jobs.csv"
)

print(jobs.shape)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

```python
sample_jobs = jobs.sample(
    10000,
    random_state=42
).reset_index(drop=True)
```

```python
job_embeddings = model.encode(
    sample_jobs["job_text"].tolist(),
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True
)
```

```python
job_embeddings.shape
```

```python
import pickle

with open(
    "job_embeddings.pkl",
    "wb"
) as f:
    pickle.dump(
        job_embeddings,
        f
    )

sample_jobs.to_pickle(
    "jobs.pkl"
)
```
