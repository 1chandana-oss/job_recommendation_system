```python
import pandas as pd

jobs = pd.read_csv("../data/postings.csv")

print(jobs.shape)

jobs.info()

jobs.head()
```

```python
keep_cols = [
    'title',
    'description',
    'company_name',
    'location',
    'formatted_experience_level',
    'formatted_work_type',
    'work_type'
]

jobs = jobs[keep_cols]

print(jobs.shape)
jobs.head()
```

```python
jobs.isnull().sum()
```

```python
jobs = jobs.dropna(
    subset=['description']
)
```

```python
jobs['company_name'] = jobs[
    'company_name'
].fillna(
    'Unknown Company'
)
```

```python
jobs[
    'formatted_experience_level'
] = jobs[
    'formatted_experience_level'
].fillna(
    'Not Specified'
)
```

```python
jobs[
    'formatted_work_type'
] = jobs[
    'formatted_work_type'
].fillna(
    'Not Specified'
)
```

```python
print(
    "Duplicates:",
    jobs.duplicated().sum()
)
```

```python
jobs = jobs.drop_duplicates()

print(jobs.shape)
```

```python
jobs[
    'description_length'
] = jobs[
    'description'
].str.len()
```

```python
jobs[
    'description_length'
].describe()
```

```python
jobs = jobs[
    jobs[
        'description_length'
    ] > 100
]
```

```python
print(jobs.shape)
```

```python
jobs = jobs.dropna(
    subset=['title']
)
```

```python
jobs.info()
```

```python
print(jobs.shape)
```

```python
jobs.to_csv(
    "clean_jobs.csv",
    index=False
)
```
