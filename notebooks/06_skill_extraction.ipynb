```python
jobs = jobs[
    [
        "title",
        "description"
    ]
]

jobs.head()
```

```python
jobs["text"] = (
    jobs["title"].fillna("") + " " +
    jobs["description"].fillna("")
)
```

```python
jobs["combined_text"] = (
    jobs["title"].fillna("").astype(str) + " " +
    jobs["description"].fillna("").astype(str)
)
```

```python


vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words="english",
    ngram_range=(2, 3),
    min_df=5,
    max_df=0.5
)

X = vectorizer.fit_transform(jobs["combined_text"])
```

```python
terms = vectorizer.get_feature_names_out()

print(len(terms))
```

```python
import numpy as np

scores = np.asarray(
    X.sum(axis=0)
).flatten()
```

```python
candidate_terms = pd.DataFrame({
    "term": terms,
    "score": scores
})

candidate_terms.head()
```

```python
candidate_terms = candidate_terms.sort_values(
    by="score",
    ascending=False
)

candidate_terms.head(100)
```
