```python
import pandas as pd

jobs = pd.read_csv("clean_jobs.csv")
```

```python
jobs['title'].value_counts().head(30)
```

```python
print(jobs.shape)

jobs.isnull().sum()
```

```python
print(
    "Unique Titles:",
    jobs['title'].nunique()
)
```

```python
top_titles = (
    jobs['title']
    .value_counts()
    .head(20)
)

top_titles
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,8))

sns.barplot(
    x=top_titles.values,
    y=top_titles.index
)

plt.title(
    "Top 20 Job Titles"
)

plt.show()
```

```python
jobs[
    'formatted_experience_level'
].value_counts()
```

```python
plt.figure(figsize=(10,5))

jobs[
    'formatted_experience_level'
].value_counts().plot(
    kind='bar'
)

plt.title(
    "Experience Level Distribution"
)

plt.show()
```

```python
jobs[
    'work_type'
].value_counts()
```

```python
plt.figure(figsize=(8,5))

jobs[
    'work_type'
].value_counts().plot(
    kind='bar'
)

plt.title(
    "Work Type Distribution"
)

plt.show()
```

```python
jobs[
    'description_length'
].describe()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    jobs[
        'description_length'
    ],
    bins=50
)

plt.title(
    "Description Length Distribution"
)

plt.show()
```
